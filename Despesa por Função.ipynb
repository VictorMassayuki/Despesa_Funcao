```python
#Estudo sobre as despesas por função da prefeitura de São José dos Campos
import pandas as pd
import seaborn as sb
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
filename = 'despesa_funcao.csv'
df = pd.read_csv(filename, engine='python', sep=';')
```

```python
df.sample(10)
```

```python
#Número de despesa por secretaria
df.groupby('Secretaria').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```python
#Número de despesa por função
df.groupby('Função').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```python
#Número de despesa por subfunção
df.groupby('Subfunção').size().sort_values().tail(20).plot(kind='barh', figsize=(10,10))
```

```python
# Gráfico com as 6 maiores despesas em saúde
df.query('Função == "SAÚDE"')['Natureza da despesa'].value_counts().sort_values().tail(6).plot(kind='barh')
```

```python
# Gráfico com as 6 maiores fontes de recursos na educação
df.query('Função == "EDUCAÇÃO"')['Fonte de recurso'].value_counts().sort_values().tail(6).plot(kind='barh')
```
